```python
import tkinter as tk

def on_radio_button_selected():
    selection = radio_var.get()
    label_result.config(text=f"Selected option: {selection}")

# Function to get user input from the entry widget
def get_user_input():
    user_input = entry_text.get()
    label_result.config(text=f"Selected option: {radio_var.get()}, User input: {user_input}")

# Create the main window
root = tk.Tk()
root.title("Radio Buttons Example")

# Create a variable to store the selected option
radio_var = tk.StringVar()

# Create radio buttons
radio_patient_number = tk.Radiobutton(root, text="Patient Number", variable=radio_var, value="Patient Number", command=on_radio_button_selected)
radio_patient_name = tk.Radiobutton(root, text="Patient Name", variable=radio_var, value="Patient Name", command=on_radio_button_selected)

# Create an entry widget for user input
entry_text = tk.Entry(root, width=30)

# Create a label to display the selected option and user input
label_result = tk.Label(root, text="Selected option: ")

# Create a button to trigger the user input retrieval
button_search = tk.Button(root, text="Search", command=get_user_input)

# Pack the widgets
radio_patient_number.pack()
radio_patient_name.pack()
entry_text.pack()
button_search.pack()
label_result.pack()

# Run the Tkinter event loop
root.mainloop()
```
